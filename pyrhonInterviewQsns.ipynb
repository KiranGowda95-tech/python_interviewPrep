Basic Interview Questions
1. Write a function that takes a list of numbers and return the sum


```python
def sum_of_list(nums):
    return sum(nums)

sum_of_list([1,2,3,4])
```

2.How do you reverse a string in python?


```python
def reverse_string(s):
    return s[::-1]

reverse_string('hello')
#Even array or list of numbers can be able to do rever by using same function
#reverse_string([1,2,3,4,5])
```

3.Write a function to check the given word is palindrome or not?

```python
def is_palindrome(word):
    return word==word[::-1]

is_palindrome('madam') #True
is_palindrome('hello') #False
```

4.How can you remove the duplicates from a list?

```python
def remove_dup(lst):
    return list(set(lst))

remove_dup([1,9,8,2,2,3,4,5,3,5,2,4])
```

6.What is a lambda function? provide an example
A lambda function is a small,anonymous function.It can take any number of arguments but can only have one expression

```python
multiply=lambda x,y:x*y
print(multiply(3,4))
```

7.Write a function that returns the n-th Fibonacci number using recurssion?

```python
def fibonacci(n):
    if n<1:
        return n
    else: 
        return fibonacci(n-1)+fibonacci(n-2)
```
