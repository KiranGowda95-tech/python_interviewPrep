3.Write a function to check the given word is palindrome or not?

```python
def is_palindrome(word):
    return word==word[::-1]

is_palindrome('madam') #True
is_palindrome('hello') #False
```

7.Write a function that returns the n-th Fibonacci number using recurssion?

```python
def fibonacci(n):
    if n<1:
        return n
    else: 
        return fibonacci(n-1)+fibonacci(n-2)
```

8. What are decorators in python and how are they used?
   Decorators provide a way to modify or extend the behavious of callable objects(like functions and methods) without permanently modifying the callable itself.
   They are often used in frameworks to add functionality to functions or methods.


```python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")

    return wrapper

@my_decorator
def say_hello():
    print("Hello!!!")

say_hello()
```

13.Explain the concept of *args and **kwargs in python?

*args allow you to pass a variable number of positional arguments to a function, while **kwargs allows you to pass a variable number of keyword arguments.

```python
def function_example(*args,**kwargs):
    for arg in args:
        print(arg)
    for key, value in kwargs.items():
        print(f"{key}={value}")

function_example(1,2,3,a=4,b=5)
```

14.What is Generator and how it is different from the list?

A generator is an iterable that yields items one at a time using a yield statement,where as list holds all items in memory.

generators are more memory-efficient for large data sets

```python
def count_up_to(n):
    count=1
    while count<=n:
        yield count
        count +=1

print("first count",count_up_to(4))
print("second count",count_up_to(1))
```

15.How can you achieve multi-threading in python?

Python has a threading module which can be used to achieve multi-threading.

```python
import threading
def print_num():
    for i in range(1,6):
        print(i)

def print_letters():
    for letter in 'abcde':
        print(letter)

t1=threading.Thread(target=print_num)
t2=threading.Thread(target=print_letters)

t1.start()
t2.start()

t1.join()
t2.join()
```

17. Describe the Global Interpreter Lock(GIL) and its implications.
The GIL is a mutex(or Lock) that allows only one thread to execute python bytecode at a time in CPython(the standard python implementation). This means that even on multi-core systems, only one thread is executed at a time.This can bea bottle neck for CPU-bound programs,but it is not generally an issue for I/O bound programs.
